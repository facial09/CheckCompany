```python
import pandas as pd
```

```python
data = pd.read_csv("all_data_202206240126.csv")
data
```

```python
def grouping(x):
    if x > 4.3 :
        return 'A'
    elif (x <=4.3) and (x > 4):
        return 'B'
    elif (3.8< x) and (x < 4):
        return 'C'
    elif (3.6< x) and (x <= 3.8):
        return 'D'
    elif (3.4< x) and (x <= 3.6):
        return 'E'
    else:
        return 'F'
```

```python
data['Group'] = data['총점'].apply(lambda x : grouping(x))
```

```python
data
```

```python
dupli = data[data.duplicated() == True].index
```

```python
data.drop(index= dupli, inplace = True)
```

```python
train = data
```

```python
train.drop(columns = ['기업추천율','ceo지지율','성장가능성'],inplace = True)
```

```python
train.drop(columns = ['name'], inplace = True)
```

```python
train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
target = ['총점']
features = ['복지및급여','워라벨','사내문화','승진기회및가능성','경영진']

model = LinearRegression()

X_train = train[features]
y_train = train[target]

model.fit(X_train,y_train)
```

```python
import pickle

with open('model.pkl','wb') as pickle_file:
    pickle.dump(model, pickle_file)
```
